**Differential equations with pyLab**

We will use the `odeint `command from the `scipy` package. Its structure is as follows:
```
n = odeint(dndt_mc,n0,time,args=(param,))
```
where

`time` is the independent variable

`n` is the dependent variable (or vector of variables)

`dndt_mc` is the name of the function giving the differential equation,

`param` contains parameter values

`time` indicates the time over which the solution is to be obtained,

`n0` represents the initial conditions.


## Code example 1: Exponential growth
Recall that the exponential growth equation is given by this expression:

$\frac{dx}{dt}=\mu x $

First in the code we need to make a block to define a function for the system of ODE equations. Then we find se solution intengrating the differential equation. For this we have to define the parameter $\mu$ we are going to pass to the function, the time vercot and the initial conditions.


````python
from matplotlib.pylab import *
from scipy.integrate import odeint


# Constitutive expression model
def dndt_mc(n,t,param):

	# Model parameters
	mu = param

	# Diferential equations
	dn = mu*n

	return dn

############################################
# SOLVE MODEL
	# Growth rate [min-1]
mu = 0.0301

param = [mu]

	# Time vector
time = arange(0.,501.,1.)

	# Initial conditions
n0 = 1.0e6

	# Solve differential equations
n = odeint(dndt_mc,n0,time,args=(param,))


time = time.reshape(501, 1)

	##################################################################
	# PLOT SIMULATION RESULTS
f1 = figure()
plot(time, (n[:, [0]])  , 'm-', label='Cells')
grid()
xlabel('Time(min)')
ylabel('Population')
title('Exponential growth')
f1.savefig('Exponential_growth.png')
````

## Code Example 2: Logistic growth equation

Recall that the logistic growth equation is given by this expression:

$\frac{dx}{dt}=\mu x \left(1-\frac{x}{K_{max}}\right)$

First in the code we need to make a block to define a function for the system of ODE equations. Then we find se solution intengrating the differential equation. For this we have to define the parameters $\mu$ and $K_{max}$ we are going to pass to the function, the time vector, and the initial conditions.

````python
from matplotlib.pylab import *
from scipy.integrate import odeint


# Constitutive expression model
def dndt_mc(n,t,param):

	# Model parameters
	mu, Kmax = param

	# Diferential equations
	dn = mu*n*(1.-n/Kmax)

	return dn

############################################
# SOLVE MODEL
	# Growth rate [min-1]
mu = 0.0301
	# Population growth capacity [cells]
Kmax = 10e8

param = [mu, Kmax]

	# Time vector
time = arange(0.,501.,1.)

	# Initial conditions
n0 = 1.0e6

	# Solve differential equations
n = odeint(dndt_mc,n0,time,args=(param,))


time = time.reshape(501, 1)

	##################################################################
	# PLOT SIMULATION RESULTS
f1 = figure()
plot(time, (n[:, [0]])  , 'm-', label='Cells')
grid()
xlabel('Time(min)')
ylabel('Population')
title('Exponential growth')
f1.savefig('Exponential_growth.png')
````
